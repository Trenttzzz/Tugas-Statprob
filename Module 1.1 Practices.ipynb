**SOAL 1**

Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus? 

**MEAN/RATA-RATA**

```python
import math
import matplotlib.pyplot as plt

DATA = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Hitung rata-rata
print(f"data : {DATA}")
data_sum = math.fsum(DATA)
data_total = len(DATA)

mean = round(data_sum / data_total)
print(f"Jumlah data : {data_sum}")
print(f"Banyak data : {data_total}")
print(f"Rata-ratanya adalah : {mean}")

# Buat Histogram
plt.hist(DATA, bins = 20, edgecolor = 'black', alpha = 1, color = 'green')

# Garis Vertikal pada mean
plt.axvline(x = mean, color = 'orange', linestyle = 'dashed', linewidth = 1, label = f'rata-rata = {mean}')

# Legend
plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')

# menampilkan plot
plt.show()
```

data : [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

Jumlah data : 1170.1

Banyak data : 11

Rata-ratanya adalah : 106

**MEDIAN / NILAI TENGAH**

```python
import math
import matplotlib.pyplot as plt

DATA = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Cari median
print(f"data : {DATA}")


# cari total data
data_total = len(DATA)

# urutkan data
sort_data = sorted(DATA)
print(f"data yang terurut : {sort_data}")
print(f"Total data : {data_total} angka")

mid_point = data_total // 2

# cek ganjil atau genap
if data_total % 2 == 0:
    print("Jumlah data genap")
    median1 = sort_data[mid_point]
    median2 = sort_data[mid_point - 1]
    median_value = (median1 + median2) / 2
else :
    print("Jumlah data genap")
    median_value = round(sort_data[mid_point], 2)

# print hasil
print(f"Nilai mediannya adalah : {median_value}")

# Buat histogram
plt.hist(DATA, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Buat sebuah garis vertikal pada median
plt.axvline(x = median_value, color ='green', linestyle = 'dashed', linewidth = 1, label = f'Median = {median_value}')

# Tambahkan label dan judul
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Frekuensi')
plt.title('Grafik Median Tinggi Badan Siswa Kelas A')

# Tambahkan legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 

# Show the plot
plt.show()
```

data : [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

data yang terurut : [98.7, 98.9, 99.7, 99.9, 100, 100.5, 101, 101.3, 102.2, 102.9, 165]

Total data : 11 angka

Jumlah data genap

Nilai mediannya adalah : 100.5

**MODUS**

```python
from collections import Counter
import matplotlib.pyplot as plt
import math

DATA = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

counter = Counter(DATA)

#ubah data ke dictionary
mode_data = dict(counter)

# mencari data yang paling banyak muncul
n = max(list(counter.values()))
print(f"angka terbanyak muncul sebanyak: {n} kali")

modus = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        modus.append(k)
print(f"Modus : {modus}")

# buat histogram
plt.hist(DATA, bins=50, edgecolor='black', alpha=0.5, color='green')

# Buat garis vertikal untuk modus 1 dan 2
for modus in modus:
  plt.axvline(x=modus, color='brown', linestyle='dashed', linewidth=1, label=f'Modus = {modus}')


# Tambahkan label dan judul
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Frekuensi')
plt.title('Modus Tinggi Badan Siswa Kelas A (cm)')

# Tambahkan legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 

# Show the plot
plt.show()
```
